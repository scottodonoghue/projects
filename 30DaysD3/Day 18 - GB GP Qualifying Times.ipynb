```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
qualy = pd.read_csv('./data/qualifying.csv')
qualy = qualy.drop(['number','constructorId','qualifyId'], axis=1)
qualy.head()                    
```

```python
races = pd.read_csv('data/races.csv')
races = races.drop(['url','circuitId','time','date','round'], axis=1)
races.head()
```

```python
drivers = pd.read_csv('data/drivers.csv')
drivers['driver_name'] = drivers.forename + " " + drivers.surname
drivers = drivers.drop(['url','nationality','dob','forename','surname','driverRef'], axis=1)
drivers.head()
```

```python
qualy_results = drivers.merge(qualy, on=['driverId']).merge(races, on=['raceId'])
qualy_results.head()
```

```python
gb_qualy = qualy_results[qualy_results.name == "British Grand Prix"]
gb_qualy.head()
```

```python
import datetime

    
def get_time_in_seconds(time_As_string):
    time_As_string = str(time_As_string)
    try: 
        t = datetime.datetime.strptime(time_As_string, '%M:%S.%f')
        minutes = int(t.strftime("%M"))
        seconds = int(t.strftime("%S"))
        milliseconds = int(t.strftime("%f"))
        return minutes*60 + seconds + (milliseconds/ 10**6)
    except ValueError as ve:
        print(f'You entered {time_As_string}, which is not a positive number.')
        return 0

q1_times = []
for i in gb_qualy.q1:
    q1_times.append(get_time_in_seconds(i))

q2_times = []
for i in gb_qualy.q2:
    q2_times.append(get_time_in_seconds(i))

q3_times = []
for i in gb_qualy.q3:
    q3_times.append(get_time_in_seconds(i))

gb_qualy['q1_seconds'] = q1_times
gb_qualy['q2_seconds'] = q2_times
gb_qualy['q3_seconds'] = q3_times
```
