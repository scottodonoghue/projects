```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Load the data into a dataframe:

```python
bikes = pd.read_csv('./data/SeoulBikeData.csv')
bikes.head(10)
```

Get a count on the number of bike hires for each day using the pandas groupby method.

```python
bike_counts = bikes.groupby(['Date']).sum().reset_index()[['Date','Rented Bike Count']]
bike_counts.columns = ['Date', 'NumBikeHires']
```

```python
bike_counts
```

Let's have a quick look at what the data will look like in a histogram:

```python
sns.histplot(data=bike_counts, x="NumBikeHires")
plt.show()
```

Finally, let's write the data to a csv file:

```python
bike_counts.to_csv('./data/seoul_daily_bike_counts.csv')
```
